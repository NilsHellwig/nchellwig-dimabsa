```python
from evaluate import evaluate_predictions, read_jsonl_file
```

```python
mode = "test-train_dev"
eval_set = "test"
# mode = "dev-train"
# eval_set = "dev"

guided = False
llm = "unsloth/gemma-3-27b-it-bnb-4bit"

VALID_COMBINATIONS = [
    (2, "eng", "restaurant"),
    (2, "eng", "laptop"),
    (2, "jpn", "hotel"),
    (2, "rus", "restaurant"),
    (2, "tat", "restaurant"),
    (2, "ukr", "restaurant"),
    (2, "zho", "restaurant"),
    (2, "zho", "laptop"),
    (3, "eng", "restaurant"),
    (3, "eng", "laptop"),
    (3, "jpn", "hotel"),
    (3, "rus", "restaurant"),
    (3, "tat", "restaurant"),
    (3, "ukr", "restaurant"),
    (3, "zho", "restaurant"),
    (3, "zho", "laptop"),
]

n_sc_map = {2: 15, 3: 10}
```

```python
for condition in VALID_COMBINATIONS:
    subtask = condition[0]
    language = condition[1]
    dataset_name = condition[2]
    n_sc = n_sc_map[subtask]

    pred_file = f"exported_predictions/{mode}/{llm.replace("/", "_")}/{n_sc}/sc_no_guided/subtask_{subtask}/pred_{language}_{dataset_name}.jsonl"
    gold_file = f"task-dataset/track_a/subtask_{subtask}/{language}/{language}_{dataset_name}_{eval_set}_task{subtask}.jsonl"
    
    predictions = read_jsonl_file(pred_file, task=subtask)
    references = read_jsonl_file(gold_file, task=subtask)
    print(f"Loaded {len(predictions)} predictions and {len(references)} subtask {subtask} references for language {language}, dataset {dataset_name}.")
    results = evaluate_predictions(references, predictions, task=subtask)
    print(f"Results for subtask {subtask}, language {language}, dataset {dataset_name}:\n", results)
    print(predictions[0])
    print(predictions[1])
    print("---")
```
